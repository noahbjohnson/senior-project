Introduction
-------------------


Tourism is often regarded as a positive force for its perceived 
effects on economic development, modernization, and globalization 
of destination populations. Much of the analysis of this issue relies 
on a number of assumptions about the nature of 'development' and its accoutrement. 
Significant relationships are often taken out of context or extrapolated inappropriately. 
The purpose of this analysis is to identify other correlations with tourist activity to 
provide further context to relationships and demonstrate the limitations of underlying assumptions.

### Processing

The data were cleaned and processed into a standardized format by subject. Extremely sparse indicators and duplicates were removed. Where needed, missing data were imputed using the [MICE](https://www.jstatsoft.org/article/view/v045i03/v45i03.pdf) method using the python package `fancyimpute`.

Statistical Methods
------------------------

### Correlation Matrixes

In each subject analysis notebook and the full analysis notebook, there are correlation matrixes provided. Using Spearman's Rank-Order Correlation, various hypothesized correlations were tested.

### Linear Regression

In the full dataset analysis, PCA is performed on each of the imputations. The resulting datasets are then used to generate linear models for tourism variables that responded well to the regression. The regression models were able to achieve a respectable coefficient of determination, but were strongly endogenous. Therefore, due to the correllated and heteroskedastic errors, instrumental variables estimation is likely the most effective tool. 

### Instrumental Variables (IV) Estimation



The first topic for analysis is where the majority of the discussions regarding the correlations of tourism are based. Economic _effects_ of tourism are often used to support further expansion of tourism activities in developing and developed nations.
